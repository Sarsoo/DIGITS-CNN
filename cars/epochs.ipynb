```python
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt

fig_dpi = 200
lw = 3
```

# Epochs
80/10/10 Split, Sigmoid, 0.01, G: 0.1

## Index
0. Epochs
1. top-1 accuracy
2. top-5 accuracy
3. last val loss
4. last val accuracy

```python
epoch_results = np.array([
    [10, 3.46, 11.74, 4.85, 5.51],
    [20, 11.98, 31.32, 3.87, 15.50],
    [30, 31.5, 59.17, 2.86, 31.92],
    [40, 39.22, 68.13, 2.68, 43.63],
    [60, 43.85, 70.11, 2.72, 47.79],
    [70, 47.62, 72.88, 2.57, 52.08],
    [80, 45.21, 73.38, 2.68, 50.43],
    [100, 45.89, 74.06, 2.74, 50.18],
    [200, 42.74, 69.24, 2.98, 47.86],
    [300, 38.91, 64.18, 3.24, 44.67],
])
```

```python
fig = plt.figure(figsize=(5, 3))
fig.set_dpi(fig_dpi)

plt.plot(epoch_results[:, 0], epoch_results[:, 1], 'x-', label="Top-1 Accuracy", lw=lw)
plt.plot(epoch_results[:, 0], epoch_results[:, 2], 'x-', label="Top-5 Accuracy", lw=lw)
plt.plot(epoch_results[:, 0], epoch_results[:, 4], 'x-', label="Final Val. Accuracy", lw=lw)

plt.ylim(0, 80)
# plt.xlim(0, 1)

plt.title('Accuracy for Varied Training Epochs')
plt.ylabel('% Accuracy')
plt.xlabel('Epochs')

plt.legend()
plt.grid()

plt.tight_layout()
plt.savefig('epochs-accuracy.png')

plt.show()
```

```python
fig = plt.figure(figsize=(5, 3))
fig.set_dpi(fig_dpi)

plt.plot(epoch_results[:, 0], epoch_results[:, 3], 'x-', label="Final Validation Loss", lw=lw)

plt.ylim(None, 5)

plt.title('Final Validation Loss for Varied Training Epochs')
plt.ylabel('Loss')
plt.xlabel('Epochs')

# plt.legend()
# plt.xscale('log')
plt.grid()

plt.tight_layout()
plt.savefig('epochs-loss.png')

plt.show()
```
